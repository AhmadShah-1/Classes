### Heap

Shape:
- All levels are full expect maybe last
- last level is filled left to right

Order:
Max-heap: every parents >= its children (max is root)
Min-heap: every parent  <= its children (min is at root)

A heap is not a sorted structure (only gurantees the parent/child relationship)


Whats the point of a heap:
A heap is for fast access to the extreme element and fast updates:
- Max/min: O(1)
- Insert: O(logn) (bubble up)

It isn't really good for "finding an arbitrary value x" as it will still be O(n) search
Instead its good for quickly changoing the largest (smallest) values when new data arrives or when priorities change.



What makes it stand out:

For a Sorted Array:
- Get Max: O(1)
- Insert:  O(n) (must shift elements)

For a Unsorted Array:
- Get max: O(n) (must scan entire array)
- Insert:  O(1)

For a Heap:
- Get Max: O(1)
- Insert: O(logn)
- Extract Max: O(logn)

extraction is O(logn), because by removing the max, the structure of the tree breaks, so you have to fix it with heapify

Heapify:
Heapify means to fix the heap property in one subtree
Essentially:
If children are heaps,
but parent might be wrong
bubble the parent down until property is restored

(Its a repair operation) O(logn)

There are two directions (Insertion, Heapify):
Bubble Up (For Insertion):
Used in (insert, increase-key)

**Moves element towards root**

Heapify Down (Maxheapify):
Used in (extract-max, building heap bottom-up)

**Moves element downwards**



Heapsort:
Heapsort is a sorting algorithm that uses a heap to repeatedly remove the largest element

Steps:
1. Turn array into max heap
2. Remove max
3. Put it at the end
4. Repeat

Heap operations:
- maxHeapify (down-heap): O(log n)
- insert: O(log n)
- extractMax: O(log n)
- increaseKey: O(log n)
- peekMax: O(1)
- build heap bottom-up: O(n)
- build heap by repeated insert (buildMaxHeap2): O(n log n)


A heap is just a binary tree but not a search tree
A heap:
- is complete
- has parent >= children
- does not have left<parent<right rule


A heap is not about searching, it is about effeciently maintaing the "current best element" while data changes
Guranteed O(nlogn) worst case sorting
In-place sorting (no extra memory)
Great as a priority queue



There are two ways to build a heap:
Method 1 (insert one by one):
In the slides this is called buildMaxHeap2  O(nlogn)

Method 2 (Bottom-Up)[More Efficient]:
Start from the last non-leaf node
Call heapify going backwards to root
More efficient as leaves are already heaps  O(n)
