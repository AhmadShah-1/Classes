```python
# To Factorial
from math import factorial
print(factorial(5))


# Using Recursion
def fact(n):
    if n == 0: 
        return 1
    else: 
        return n * fact(n-1)

print(fact(5))
```

```python
# Returns the legth of a list using recursion
def lstLength(lst):
    if lst == []:
        return 0
    else:
        return lstLength(lst[1:]) + 1

print(lstLength([1,2,3,4,5]))


# Returns the sum of a list using recursion
def sumList(lst):
    if lst == []:
        return 0
    else:
        return lst[0] + sumList(lst[1:])

print(sumList([1,2,3,4,5]))


# Returns the concatenation of a list using recursion
def listConcat(lst):
    if len(lst) == 1:
        return lst[0]
    else:
        return lst[0] + listConcat(lst[1:])

print(listConcat(['a','b','c']))
```

```python
# Returns the sum of a nested list using recursion
def nestedSum(lst):
    lstSum = 0
    for i in lst:
        if not isinstance(i, list):
            lstSum += i
        else:
            lstSum += nestedSum(i)
    return lstSum

print(nestedSum([1,2,[3,4],[5,6]]))
```

```python
# Permutations of a sequence
def permute1(seq):
    if not seq:
        return [seq]  # base case. Empty sequence - if list is empty, return empty list
    else:
        result = []
        for i in range(len(seq)):
            rest = seq[:i] + seq[i+1:]   # All elements except the current
            for x in permute1(rest):     # Recursively permute the rest
                result.append(seq[i:i+1] + x)  # Add node at front; seq[i:i+1] preserves the original sequence type
                
        """
        1st iteration: 
        i = 0, pick seq[0] = 1; rest = seq[:0] + seq[1:] = [2, 3]
            permute [2, 3] 
                    pick 2, permute [3]
                            -> [3]
                            combine [2] and [3] -> [2, 3]
                    pick 3, permute [2] 
                            -> [2]
                            combine [3] and [2] -> [3, 2]
        combine with [1] with [2, 3] -> [1, 2, 3]
        combine with [1] with [3, 2] -> [1, 3, 2] 
        """           

        """
        2nd iteration: i = 1, pick seq[1] = 2, rest = seq[:1] + seq[2:] = [1, 3]
        """

        """
        3rd iteration: i = 2, pick seq[2] = 3, rest = seq[:2] + seq[3:] = [1, 2]
        """
 
                #print(result)
        return result
        
permute1([1, 2, 3])                                                   
```

```python
# Using Generator Function
def permute2(seq):
    if len(seq) == 1:    
        yield seq   # Base case
    else:
        for i in range(len(seq)):
            current = seq[i:i+1]
            rest = seq[:i] + seq[i+1:]    
            for x in permute2(rest): 
                #print(seq[i:i+1] + x)
                yield current + x 

p2 = permute2([1, 2, 3])                                                  
```

```python
import sys

seq = [1, 2, 3, 4, 5, 6, 7, 8, 9]
"""
sys.getsizeof() only measures the container object itself, not all internal allocations.
So itâ€™s good for relative comparison, not total memory.
"""

print("Size of function-based:", sys.getsizeof(permute1(seq)), "bytes")  
# Builds up n stack frames (one per recursive call). Memory grows roughly O(n).

print("Size of generator-based:", sys.getsizeof(permute2(seq)), "bytes")
# Stores only the state of the iteration (the generator frame). Memory usage is O(1) -- constant
```
