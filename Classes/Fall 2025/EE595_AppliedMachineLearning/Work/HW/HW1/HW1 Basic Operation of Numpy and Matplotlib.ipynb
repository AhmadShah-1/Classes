## Question 4
For the previous generated data, consider each row of the array as an individual data point, with the two columns representing its distinct features. Identify the data points with the smallest and largest sum of features. Mark them in red and blue in the figure respectively.

Hint: You can compute the sum of each row and store it in an array named "sums". Then find the minimum or maximum index of this array.

```python
# Your Code
sums = []
for array in Array1:
    sums.append([array[0] + array[1], array[0], array[1]])

min_row = min(sums, key=lambda x: x[0])
max_row = max(sums, key=lambda x: x[0])


plt.scatter(Array1[:, 0], Array1[:, 1], c="black", label="Data Points")
plt.plot(min_row[1], min_row[2], "ro", label="Smallest Sum")
plt.plot(max_row[1], max_row[2], "bo", label="Largest Sum")

plt.title("Scatter Plot of Randomly Generated Data from a uniform distribution over [0, 1]")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.2), ncol=3)
plt.gca().xaxis.set_major_locator(MultipleLocator(0.1))



plt.show()


print(max_row)
print(min_row)
print(sums)
```

## Question 5
Add two lines representing the mean of the Feature 1 and the 25% quantile of the Feature 2 seperately. Add the legend of the two lines.

```python
# Your Code
elements = len(Array1)
mean = 0
quantile_calc = []

for array in Array1:
    mean += array[0]
    quantile_calc.append(array[1])

mean /= elements
quantile_25 = np.quantile(quantile_calc, 0.25)

plt.axvline(mean, color="green", linestyle="-", label="Mean of Col 1")
plt.axhline(quantile_25, color="orange", linestyle="-", label="25% Quantile of Col 2")
plt.scatter(Array1[:, 0], Array1[:, 1], c="black", label="Data Points")

plt.title("Scatter Plot of Randomly Generated Data from a uniform distribution over [0, 1]")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.2), ncol=3)
plt.gca().xaxis.set_major_locator(MultipleLocator(0.1))

plt.show()


print(mean)
print(quantile_25)
```

```python
# 1
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set_style("whitegrid")
import numpy as np 
from sklearn.metrics import mean_squared_error


noise_scale = 100 
number_of_samples = 50 
x =  30*(np.random.rand(number_of_samples, 1) - 0.5) 
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 

plt.plot(x,y,'ro') 
```

```python

x = x.flatten()
y = y.flatten()

m = 3  
coeffs = np.polyfit(x, y, m)
p = np.poly1d(coeffs)

x_line = np.linspace(min(x), max(x), 200)
y_line = p(x_line)

plt.scatter(x, y, c="red", label="Noisy data")
plt.plot(x_line, y_line, c="blue", label=f"Fitted polynomial (m={m})")
plt.legend()
plt.title("Polynomial Fit with m=3")
plt.show()
```

```python
# 2
orders = [1, 2, 3, 4, 5, 6, 7, 8]
mse_list = []

for m in orders:
    coeffs  = np.polyfit(x, y, m)
    p = np.poly1d(coeffs)
    output = p(x)
    mse = np.mean((y - output)**2)
    mse_list.append(mse)


lowest_index = mse_list.index(min(mse_list))
best_order = orders[lowest_index]

print(mse_list)
print(f"Order {lowest_index} has the lowest MSE")

plt.plot(orders, mse_list, marker="o")
plt.xlabel("Polynomial Order")
plt.ylabel("MSE")
plt.title("MSE vs Polynomial Order")
plt.show()
```

```python
# 3
best_m = 7  

noise_scales = [200, 300, 400, 600, 800, 1000]

for noise in noise_scales:
    x = 30 * (np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise * np.random.randn(number_of_samples, 1)
    x = x.flatten()
    y = y.flatten()

    coeffs = np.polyfit(x, y, best_m)
    p = np.poly1d(coeffs)

    x_line = np.linspace(min(x), max(x), 200)
    y_line = p(x_line)
    
    plt.figure(figsize=(6,4))
    plt.scatter(x, y, c="red", label="Noisy data")
    plt.plot(x_line, y_line, c="blue", label=f"Poly fit (m={best_m})")
    plt.title(f"Noise scale = {noise}")
    plt.legend()
    plt.show()



    '''
    With a lower noise level from 200 to 300 the data is spreda around the curve, and the polynomial of 7 fits the relationship very well
    Once the noise starts to increase from 400 to 600, the points are more scattered, however the curve still follows the trend of the data
    Once we reach a noise of 800 it becomes very noticable that the fit of the curve starts to break down, following the trend of the noise, reducing the efficacy of the coefficients.
    As the noise grows the model fits itself towards the noise and random variations it detects rather than the true data trend.
    The higher order polynomial of 7 does introduce flexibility in terms with fitting against noise, however at higher noise levels, this reduces generalization.
    '''
```

```python
# 4
samples = [40, 30, 20, 10]
noise_scale = 100  

for n in samples:
    x = 30 * (np.random.rand(n, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale * np.random.randn(n, 1)
    x = x.flatten()
    y = y.flatten()

    coeffs = np.polyfit(x, y, best_m)
    p = np.poly1d(coeffs)

    x_line = np.linspace(min(x), max(x), 200)
    y_line = p(x_line)
    plt.figure(figsize=(6,4))
    plt.scatter(x, y, c="red", label="Noisy data")
    plt.plot(x_line, y_line, c="blue", label=f"Poly fit (m={best_m})")
    plt.title(f"Number of samples = {n}")
    plt.legend()
    plt.show()


'''
When 40 samples are present, the curve still remains intact, this is due to the fact that there are still enough values present across the plot that it contrains the progression of the curve
When we reach 20 samples, we can noticably determine there is some deviation from the trend, however there is still enough values present to justify a good precision for the curve.
However, once we approach 10 samples, it is clear that the curve is no longer intact, and that it is highly unstable, trying to fit towards the few points available, leading to overfitting,
The more samples that are present in the dataset, the more stable the parameters will become
The issue with having fewer samples is that it leads to overfitting, where small changes or outliers will have a very noticable effect on the coefficients.
'''
```
