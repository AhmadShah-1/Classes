```python
class Car:
    wheels = 4  # class attribute (shared default)

    def __init__(self, name):
        self.name = name

    @classmethod
    def change_wheels(cls, n):
        cls.wheels = n

    @classmethod
    def show_wheels(cls):
        print(f"[CLASS] {cls.__name__}.wheels = {cls.wheels}")

    def show(self):
        print(f"[INSTANCE] {self.name}: wheels = {self.wheels}")


# ---- demo ----
Car.show_wheels()

a = Car("A")
b = Car("B")
a.show()
b.show()

print("\nChange wheels via classmethod...\n")
Car.change_wheels(3)

Car.show_wheels()
a.show()
b.show()

print("\nOverride wheels on ONE instance only...\n")
a.wheels = 10  # instance attribute overrides class attribute lookup
a.show()
b.show()
Car.show_wheels()
```

```python
class Person:
    def __init__(self, name, pay):
        self.name = name
        self.pay = pay

    def giveRaise(self, percent):
        self.pay = int(self.pay * (1 + percent))

    def __str__(self):
        return f"{self.name}'s salary is {self.pay:.1f}"

class Manager(Person):
    def giveRaise(self, percent, bonus = .10):
        super().giveRaise(percent + bonus)

p = Person('Sue', 80)
m = Manager('Pat', 100)

for obj in [p, m]:
    obj.giveRaise(.10)
    print(obj)
```
