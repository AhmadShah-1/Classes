```python
'''
1.
Composition is when a class is made using one or more objects of another class. 
It contains the objects of the other class and can use their functions.
'''

# Composition Example (Door in game)
class Door:
    def __init__(self, material):
        self.material = material

    def open(self):
        print("Opening the door")

class House:
    def __init__(self, door_material):
        # Composition
        self.door = Door(door_material)

    def enter(self):
        self.door.open()

house1 = House("Wood")
house1.enter()
```

```python
'''
Delegation is when an object of a class is passed to another class to use it to complete a task. 
Such as a calculcator object passing the calculation to a calculator class.
'''
# Delegation Example (Calculator)
class mathfunc:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

class Calculator:
    def __init__(self):
        # Delegation
        self.calc = mathfunc()

    def computation(self, operation, a, b):
        print("Calculating...")

        if operation == "add":
            return self.calc.add(a, b)
        elif operation == "subtract":
            return self.calc.subtract(a, b)
        else:
            return "Invalid operation"

calc1 = Calculator()
print(calc1.computation("add", 1, 2))
print(calc1.computation("subtract", 1, 2))
```
