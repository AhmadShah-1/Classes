```python
# useMyModule.py
'''
This program imports the myModule file, and uses matplotlib to display each student and their grades using a bar plot
'''

import myModule
import matplotlib.pyplot as plt

# Extract Data from myScores() func, and initialize students and averages lists
data = myModule.myScores()
students = []
averages = []

# Parse data, seperating into appropriate ordered categories
for student, avg in data.items():
    students.append(student)
    averages.append(avg)
    print(student + " " + str(avg))

# Plot Results
plt.bar(students, averages)
plt.xlabel('Students')
plt.ylabel("Average Grade")
plt.title("Students and Average Grade")
plt.show()
```
