```python
'''
1.
This program creates a Point class that can be used to create points in a 2D space.
'''

class Point():
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overload subtration operator
    def __sub__(self, second):
        return Point(self.x - second.x, self.y - second.y)

    # Overload string operator
    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
p3 = p1 - p2
print(p3)
```

```python
'''
2.
This program creates a circle with a given radius and calculates the area and circumference.
'''

class Circle:
    def __init__(self, radius):
        self.radius = radius

    # Overload getattr operator
    def __getattr__(self, name):
        # if name is area, return the area of the circle
        if name == 'area':
            return 3.14 * self.radius * self.radius
        # if name is circumference, return the circumference of the circle
        elif name == 'circumference':
            return 2 * 3.14 * self.radius
        raise AttributeError(name)
 

c = Circle(3)
print(c.area)            
print(c.circumference)
```

```python
'''
3.
Encapsulation is the process of hiding certain attributes and methods to ensure data is protected
(Data the user doesn't need to see)
Polymorphism allows objects of different classes to be accessible from the same interface.
(Same method name can be used for different classes)
'''

# Example of encapsulation
class student:
    def __init__(self, name, grade):
        # Private attributes (Do not access directly)
        self.__name = name
        self.__grade = grade

    # Getter methods (Access private attributes)
    def get_name(self):
        return self.__name
    
    def get_grade(self):
        return self.__grade
    
    def set_grade(self, grade):
        self.__grade = grade

# Create an instance of the student class
student1 = student("John", 95)
print(student1.get_name())
print(student1.get_grade())

# Example of polymorphism
class Animal:
    # Base Method (Will be overridden in the child classes)
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Woof"

class Cat(Animal):
    def sound(self):
        return "Meow"

# Create instances of the Dog and Cat classes
dog = Dog()
cat = Cat()

print(dog.sound())
print(cat.sound())
```

```python
'''
4.
This program creates a class for undergraduate and graduate students.
'''

# Undergraduate class
class undergraduate:
    # Constructor
    def __init__(self, name, grade, classes):
        self.__name = name
        self.__grade = grade
        self.__classes = classes

    def get_name(self):
        return self.__name
    
    def get_grade(self):
        return self.__grade
    
    def set_grade(self, grade):
        self.__grade = grade

    def get_classes(self):
        return self.__classes

    def set_classes(self, classes):
        self.__classes = classes

class graduate(undergraduate):
    # Constructor 
    def __init__(self, name, grade, classes, thesis):
        # Call the constructor of the undergraduate class
        super().__init__(name, grade, classes)
        self.__thesis = thesis

    # Getter method
    def get_thesis(self):
        return self.__thesis

    # Setter method
    def set_thesis(self, thesis):
        self.__thesis = thesis

student1 = undergraduate("Leonard", 95, ["Math", "Science", "English"])
student2 = graduate("Jacob", 85, ["Math", "Science", "English"], "Thesis on Electrical Engineering")

print(student1.get_name())
print(student2.get_name())

print(student1.get_classes())
print(student2.get_classes())

print(student2.get_thesis())
```
