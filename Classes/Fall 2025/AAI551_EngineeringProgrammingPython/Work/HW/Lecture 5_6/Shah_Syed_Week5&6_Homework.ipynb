```python
'''
Q1
The function my_sum takes as input a list, and solves for the sum of all elements within the list recursively
'''
def my_sum(lst):
    if len(lst) == 0:   # Base Case (indicate when to stop)
        return 0
    else:
        return lst[0] + my_sum(lst[1:])  # Add the first element in the list and recursively solve

lst = [1, 2, 3, 4]
print(my_sum(lst))
```

```python
'''
Q2
This program uses map to take the users input for two numbers and return the difference using a lambda function to perform the operation
'''

x, y = map(int, input("Enter Two Numbers separated by a space: ").split())  # Use split to split the string by whitespace, then convert to int
lambda_diff = lambda a, b : a - b

print(lambda_diff(x, y))
```

```python
'''
Q3
This program takes a name and age as input and prints out a statement using an f string
I opted for using lambda and map to gain a better understanding of using it
'''

def greeting(name, age):
    # Check if variables are valid
    if not isinstance(name, str):
        return "Name must be a string"
    if not isinstance(age, float) and not isinstance(age, int):
        return "Age must be a float"

    return f"Hi, my name is {name} and I am {age} years old"


# User Input
name_age = input("Enter your name and age seperated by a space \n if you wish to run a test, leave the entry blank: \n")


if len(name_age) == 0:
    names = ["Jake", "Syed", "Ahmad", 4]
    ages = [22, 21.4, "21", 33]

    # Map names and age and feed it to the function
    list_outputs = list(map(greeting, names, ages))

    # Two different ways of performing the numbering of each iteration
    # final_output = list(number + list_outputs[item] for item, number in enumerate(list_outputs)) 
    final_output = lambda x: [str(number+1) + ". " + item for number, item in enumerate(x)]

    # Apply print() function on every value
    lambda_print = lambda x: list(map(print, x))

    lambda_print(final_output(list_outputs))
else:
    name, age = name_age.split()
    age = float(age)

    print(greeting(name, age))
```

```python
'''
Q5
This function implements the Ackermann function for m and n recursively
'''
def my_ackermann(m, n):
    # If statements for each case
    if m == 0:
        return n + 1
    elif m > 0 and n == 0:
        return my_ackermann(m-1, 1) # Call function again until base case is reached
    elif m > 0 and n > 0:
        return my_ackermann(m-1, my_ackermann(m, n-1))  # Call function recursively

print(my_ackermann(0, 1))
print(my_ackermann(1, 0))

print(my_ackermann(1, 1))
print(my_ackermann(1, 2))
print(my_ackermann(2, 3))
```
