An exception is an error that is thrown during a program, that allows it to be caught and handled appropriately, 
preventing the program from crashing. 

```python
'''
This code demonstrates exception handling for try and except
'''

def read_from_file(filename):
    try:
        with open(filename, 'r') as file:
            # Read the first line of the file and convert it to an integer
            value = int(file.read().strip())
            print("Integer is ", value)

    # If file not found, print an error message
    except FileNotFoundError:
        print("File not found")

    # If the value returned is not an integer, print an error message
    except ValueError:
        print("Not an integer value")

read_from_file("file.txt")
```

```python
'''
This code shows how to use raise using a login check to demonstrate it
'''

def login(username, password):
    # Check if the username and password are correct
    if username != "admin" or password != "12345":
        raise ValueError("Invalid username or password")
    print("Login successful")

try:
    login("admin", "12345")
except ValueError as e:
    print(e)
```

```python
'''
This code demonstrates assert for validation
'''

def avg(values):
    # Check if the list is empty
    assert len(values) > 0, "The list is empty"
    return sum(values) / len(values)

avg([])
```
