```python
def myRev(L):
    if L == []:
        return []
    else:
        return myRev(L[1:]) + [L[0]]

print(myRev([1,2,3,"abc"]))
```

```python
def isPrime(n):
    if n == 0 or n == 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True




```

```python
from collections import deque

L= [1,2,3,4,5]
queue = deque(L)
for i in range(len(L)):
    temp = queue.popleft()  # Remove from left
    queue.append(temp)     

print(queue)
```
