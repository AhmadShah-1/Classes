```python
import numpy as np
import scipy.io as io
hw3_1 = io.loadmat("hw3.mat")['hw3_1']
hw3_2_1 = io.loadmat("hw3.mat")['hw3_2_1']
hw3_2_2 = io.loadmat("hw3.mat")['hw3_2_2']
```

```python
X = hw3_1               # shape (2, 100)
d, N = X.shape          # d = 2, N = 100

print(d)
print(N)
```

```python
# label vector: first 50 = 1, next 50 = 2
labels = np.array([1]*50 + [2]*50)

# p_0 as specified in Step1
rho = 0.5
```

```python
def gauss(X_subset):
    mu = np.mean(X_subset, axis=1, keepdims=True) 
    diffs = X_subset - mu
    n = X_subset.shape[1]
    Sigma = (diffs @ diffs.T) / n   
    return mu, Sigma

# Assign each sample to one of the two guassians 
idx1 = (labels == 1)
idx2 = (labels == 2)

X1 = X[:, idx1]
X2 = X[:, idx2]

mu1, Sigma1 = gauss(X1)
mu2, Sigma2 = gauss(X2)

rho = idx1.sum() / N   
```

```python
def mvn_logpdf(x, mu, Sigma):
    d = x.shape[0]
    diff = x - mu
    sign, logdet = np.linalg.slogdet(Sigma)
    if sign <= 0:
        sign, logdet = np.linalg.slogdet(Sigma)
    inv = np.linalg.inv(Sigma)
    return -0.5 * (d * np.log(2*np.pi) + logdet + diff.T @ inv @ diff)



def reassign_labels(X, mu1, Sigma1, mu2, Sigma2):
    N = X.shape[1]
    new_labels = np.zeros(N, dtype=int)
    for n in range(N):
        x = X[:, n]
        l1 = mvn_logpdf(x, mu1.ravel(), Sigma1)
        l2 = mvn_logpdf(x, mu2.ravel(), Sigma2)

        # Assign to Gaussian with larger likelihood to fill out new_labels
        new_labels[n] = 1 if l1 >= l2 else 2
    return new_labels
```

```python
history = []

labels = np.array([1]*50 + [2]*50)  # reset initial labels
rho = 0.5

# ITerate for 5 times
for it in range(5):
    idx1 = (labels == 1)
    idx2 = (labels == 2)
    X1 = X[:, idx1]
    X2 = X[:, idx2]

    mu1, Sigma1 = gauss(X1)
    mu2, Sigma2 = gauss(X2)
    rho = idx1.sum() / N

    history.append((mu1.copy(), Sigma1.copy(), mu2.copy(), Sigma2.copy(), rho, labels.copy()))

    # Reassign labels using likelihoods
    labels = reassign_labels(X, mu1, Sigma1, mu2, Sigma2)

# Final parameters after 5 iterations
mu1_f, Sigma1_f, mu2_f, Sigma2_f, rho_f, labels_f = history[-1]
```
